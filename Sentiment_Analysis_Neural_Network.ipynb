```
import numpy as np
import tensorflow as tf
from tensorflow import keras
from keras.datasets import imdb
from keras.preprocessing.sequence import pad_sequences
VOCAB_SIZE = 10000
MAX_LEN = 250
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)
print(f"Loaded {len(x_train)} training reviews and {len(x_test)} test reviews.")

#Padding is required to make the sequences to a same length so that they can be processed by the neural network.

print(f"Padding sequences to a max length of {MAX_LEN} words...")
x_train_padded = pad_sequences(x_train, maxlen=MAX_LEN, padding='post', truncating='post')
x_test_padded = pad_sequences(x_test, maxlen=MAX_LEN, padding='post', truncating='post')

EMBEDDING_DIM = 16
model = keras.Sequential([
    keras.layers.Embedding(VOCAB_SIZE, EMBEDDING_DIM, input_length=MAX_LEN),
    keras.layers.GlobalAveragePooling1D(),
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
model.fit(x_train_padded,y_train, epochs=45,batch_size=512,validation_data=(x_test_padded, y_test),verbose=1)

word_index = imdb.get_word_index()
word_index = {k:(v+3) for k,v in word_index.items()}
word_index["<PAD>"] = 0
word_index["<START>"] = 1
word_index["<UNK>"] = 2

def preprocess_text(text):
    tokens = text.lower().split()
    encoded_tokens = [word_index.get(word, 2) for word in tokens]
    padded_input = pad_sequences([encoded_tokens], maxlen=MAX_LEN, padding='post', truncating='post')
    return padded_input

print("\n--- Sentiment Analysis CLI ---")
print("Enter a sentence to analyze, or type 'quit' to exit.")

while True:
    user_input = input("\nYour sentence: ")
    if user_input.lower() == 'quit':
        break

    if not user_input.strip():
        print("Please enter some text.")
        continue

    processed_input = preprocess_text(user_input)
    prediction = model.predict(processed_input)
    score = prediction[0][0]
    if score > 0.6:
        sentiment = "Positive ğŸ˜„"
    elif score < 0.4:
        sentiment = "Negative ğŸ˜"
    else:
        sentiment = "Neutral ğŸ˜"

    print(f"Sentiment: {sentiment}")
    print(f"Confidence Score: {score:.4f}")
```
